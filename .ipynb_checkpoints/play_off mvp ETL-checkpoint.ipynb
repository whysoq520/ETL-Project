```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
mvp_file ="NBA Finals and MVP.xlsx"
mvp_df = pd.read_excel(mvp_file)
mvp_df.head()
```

```python
df_drop=mvp_df.iloc[6:]
df_drop.head()
```

```python
new_mvp_df =df_drop[['MVP Frist_Name', 'MVP Last_Name','Year','MVP Position','MVP Team','MVP Nationality','MVP status']].copy()
new_mvp_df.head()
```

```python
mvp = new_mvp_df.rename(columns={ 
                                 "MVP Frist_Name":"First_name",
                                 "MVP Last_Name":"Last_name",
                                 "MVP Position": "Position",
                                 "MVP Team": "Team",
                                 "MVP Nationality":"Nationality",
                                 "MVP status":"MVP_status"
                                })
mvp.head()
```

```python
team=pd.read_csv("Resources/TeamDetails.csv")
team.head()
```

```python
mvp_team =mvp.merge(team,left_on ="Team", right_on ="team_name", how ="inner")
mvp_team.head()
```

```python
mvp_team.columns
```

```python
mvp_df = mvp_team[['First_name', 'Last_name', 'Year', 'Position', 'Team', 'Nationality',
       'MVP_status', 'team_id']].copy()
mvp_df.head()
```

```python
mvp_final = mvp_df.rename(columns={ "team_id":"Team_id"})
mvp_final.head()
```

```python
mvp_final.to_csv("Resources/Play_off MVP.csv")
```
