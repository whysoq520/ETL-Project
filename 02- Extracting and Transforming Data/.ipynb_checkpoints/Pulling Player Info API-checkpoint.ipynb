# Extracting the players information from NBA Sense 

```python
# defining the url of the API for the year 2016

url= 'http://stats-prod.nba.com/wp-json/statscms/v1/synergy/player//?category=Transition&season=2016&seasonType=Reg&names=offensive&limit=500'
```

```python
result=requests.get(url).json()
```

```python
result
```

```python
# Appending player information to the different list: 2016


info = ['PlayerIDSID', 'PlayerFirstName','PlayerLastName','TeamIDSID', 'P']
player_id=[]
player_first_name = []
player_last_name = []
player_team_id = []
player_position = []


for player in result['results']:
    player_id.append(player['PlayerIDSID'])
    player_first_name.append(player['PlayerFirstName'])
    player_last_name.append(player['PlayerLastName'])
    player_team_id.append(player['TeamIDSID'])
    player_position.append(player['P'])
    
```

```python
# defining the url of the API for the year 2017

url1= 'http://stats-prod.nba.com/wp-json/statscms/v1/synergy/player//?category=Transition&season=2017&seasonType=Reg&names=offensive&limit=500'

result1=requests.get(url1).json()
```

```python
# Appending player information to the different list: 2017



for player in result1['results']:
    player_id.append(player['PlayerIDSID'])
    player_first_name.append(player['PlayerFirstName'])
    player_last_name.append(player['PlayerLastName'])
    player_team_id.append(player['TeamIDSID'])
    player_position.append(player['P'])
```

```python
# defining the url of the API for the year 2018

url2= 'http://stats-prod.nba.com/wp-json/statscms/v1/synergy/player//?category=Transition&season=2018&seasonType=Reg&names=offensive&limit=500'

result2=requests.get(url2).json()
```

```python
# Appending player information to the different list: 2018



for player in result2['results']:
    player_id.append(player['PlayerIDSID'])
    player_first_name.append(player['PlayerFirstName'])
    player_last_name.append(player['PlayerLastName'])
    player_team_id.append(player['TeamIDSID'])
    player_position.append(player['P'])
```

```python
# defining the url of the API for the year 2019

url3= 'http://stats-prod.nba.com/wp-json/statscms/v1/synergy/player//?category=Transition&season=2019&seasonType=Reg&names=offensive&limit=500'

result3=requests.get(url3).json()
```

```python
# Appending player information to the different list: 2019



for player in result3['results']:
    player_id.append(player['PlayerIDSID'])
    player_first_name.append(player['PlayerFirstName'])
    player_last_name.append(player['PlayerLastName'])
    player_team_id.append(player['TeamIDSID'])
    player_position.append(player['P'])
```

```python
# Creating a dataframe to hold the list information in the columns (one for each list)

players_info= pd.DataFrame()
```

```python
# Creating the columns and uploading the list in the dataframe
players_info['player_id']  = player_id
players_info['first_name']  = player_first_name
players_info['last_name']  = player_last_name
players_info['team_id']  = player_team_id
players_info['position']  = player_position

```

```python
players_info.head()
```

```python
#Export file as a CSV, without the Pandas index, but with the header
players_info.to_csv("Players Information_API.csv", index=False, header=True)
```

# Transforming the dataframe to hold only unique player's ID, first and last name

```python
# Creating a data frame with only players's ID, firsty and last name fields

player_tableinfo = players_info.iloc[:,0:3]
player_tableinfo.head()
```

```python
# Checking total number of records
len(player_tableinfo['player_id'])
```

```python
# Removing duplicates (player_id) from the dataframe and creating a new dataframe
players_unique = player_tableinfo.drop_duplicates(subset=['player_id'], keep='first')
len(players_unique['player_id'])
```

```python
#Export file as a CSV, without the Pandas index, but with the header
players_unique.to_csv("\Resources\Players Table.csv", index=False, header=True)
```
