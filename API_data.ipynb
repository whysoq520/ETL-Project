```python
# Import the dependencies
import requests
import json
import pandas as pd
```

```python
team_profile=[]
team_details=[]
h_of_fame = []
teamid_list = []
```

```python
# Make API requests for each team
for team_id in range(1610612737, 1610612767):
    url= (f'https://stats.nba.com/feeds/teams/profile/{team_id}_TeamProfile.js')
    result=requests.get(url).json()
    team_profile.append(result)
```

```python
# Print the result
print(team_profile)
```

```python
# Get team details from the API
i=0
for i in range(len(team_profile)):
    team_details.append(team_profile[i]['TeamDetails'][0]['Details'])
    i=i+1
```

```python
print(team_details)
```

```python
# Get all the team_ids
i=0
for i in range(len(team_profile)):
    teamid_list.append(team_profile[i]['TeamDetails'][0]['Details'][0]["Team_Id"])
    i=i+1
```

```python
print(teamid_list)
```

```python
# Initialize a dataframe for team details
df_teamdetails = pd.DataFrame(team_details[0])
```

```python
# Add all the teams to the dataframe
i=1
for i in range(len(team_details)):
    df_teamdetails = df_teamdetails.append(team_details[i])
    i=i+1
```

```python
# Reorganize dataframe
df_teamdetails = df_teamdetails[["Team_Id","Abbreviation","Nickname","YearFounded", "YearActiveTill", "ArenaCapacity"]]
df_teamdetails
```

```python
# Rename the columns
df_teamdetails = df_teamdetails.rename(columns={"Team_Id":"team_id","Abbreviation":"abbreviation","Nickname":"team_name","YearFounded":"year_founded", "YearActiveTill":"yearactive_till", "ArenaCapacity":"arena_capacity"})
df_teamdetails.head()
```
