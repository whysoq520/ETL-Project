```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
mvp_file ="NBA Finals and MVP.xlsx"
mvp_df = pd.read_excel(mvp_file)
mvp_df.head()
```

```python
new_mvp_df = mvp_df[['MVP Name', 'Year','MVP Position','MVP Team','MVP Nationality','MVP status']].copy()
new_mvp_df.head()
```

```python
mvp = new_mvp_df.rename(columns={"Year": "season",
                                 "MVP Name": "MVP_name",
                                 "MVP Position": "MVP_Position",
                                 "MVP Team": "MVP_team",
                                "MVP Nationality":"MVP_nationality",
                                 "MVP status":"MVP_status"
                                })
mvp.head()
```
