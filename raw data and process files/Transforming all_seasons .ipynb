```python
import pandas as pd
```

```python
nba_file = "all_seasons.csv"
nba_df = pd.read_csv(nba_file)
nba_df.head()
```

```python
nba_df[['first_name','last_name']] = nba_df['player_name'].loc[nba_df['player_name'].str.split()
                                                               .str.len() == 2].str.split(expand=True)

nba_df.head()
```

```python
# Create a filtered dataframe from specific columns
player_columns = ["first_name", "last_name", "age", "player_weight", "player_height", "college","country", "season", "gp", 
               "pts", "reb", "ast"]

nba_transformed = nba_df[player_columns].copy()
```

```python
nba_transformed.head()
```

```python
# Rename the column headers
final_nba_df = nba_transformed.rename(columns={"gp": "games_played",
                                                  "pts": "points",
                                                  "rb": "rebounds",
                                                  "ast": "assistance_pergame"})
```
